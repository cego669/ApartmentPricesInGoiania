## Price of the apartments

Here we will construct an histogram so we can see how the prices of the apartments are distributed. Note that, as the prices are strictly positive and have a range of values that can easily explode, it's better to visualize them using a function transformation like `np.log10`, that takes the logarithm on the base 10 of the values:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(np.log10(data["price"]), bins=350)
plt.xlabel("log10(apartment prices)")
plt.axvline(np.log10(data["price"].quantile(0.001)), linestyle="--", label="0.1% quantile")
plt.axvline(np.log10(data["price"].quantile(0.999)), linestyle="--", label="99.9% quantile")
plt.legend()
plt.show()
```

Note that there are prices strangely low (less than 10000 reais). In Brazil, properties like apartments normally cost over 100000 reais. Let's see how many instances present prices lower than 10000 reais:

```python
data[data["price"] <= 10**5]
```

Only 3 instances. That's interesting. Let's read the complete description for these apartments so we can try to understand why these prices are so low:

```python
data[data["price"] <= 10**5]["complete_description"][75]
```

```python
data[data["price"] <= 10**5]["complete_description"][7801]
```

```python
data[data["price"] <= 10**5]["complete_description"][9259]
```

We can see additionaly that the complete description is exempt of a pattern for the strings. Let's fix this:

```python
from unidecode import unidecode
from string import punctuation

def temp(x):
    x = unidecode(" ".join(x.lower().split()))
    for p in punctuation:
        x = x.replace(p, " ")
    for n in "0123456789":
        x = x.replace(n, " ")
    return " " + " ".join(x.split()) + " "

data["complete_description"] = data["complete_description"].apply(temp)
```

```python
data["complete_description"][9259]
```

It's a lot better now. The strings don't have accents.

From the 3 complete description above, we were able to see that tha main cause for the low prices are the fact that the apartments come from an "ágio". If you google it, you will se that an apartment "ágio" "*is the sale of a property that is still being financed, where the seller gives up the right to pay the remaining installments to a new buyer. The seller may be the original buyer of the property who no longer wants to continue paying the installments, or it may be someone who purchased the property and then sold it out of necessity*". There are probably still many installments to be paid for these apartments, which justifies the low sale price.

Let's see if there are other apartments that are being sold as an "ágio":

```python
def temp(x):
    for w in [" agio ", " agil ", " agiu ", " ajil ", " ajio ", " ajo "]:
        if w in x:
            return True
        else:
            return False

data[data["complete_description"].apply(temp)]
```

There are more than 250 instances. I would just drop all the "ágio" instances, but it's better to just drop the rows with prices lower than 10000 reais.

```python
data = data[data["price"] > 10**5].copy()
```

Now we can see the new distribution for the log(price):

```python
plt.hist(np.log10(data["price"]), bins=350)
plt.xlabel("log10(apartment prices)")
plt.axvline(np.log10(data["price"].quantile(0.001)), linestyle="--", label="0.1% quantile")
plt.axvline(np.log10(data["price"].quantile(0.999)), linestyle="--", label="99.9% quantile")
plt.legend()
plt.show()
```

## Latitude and longitude

It is useful to visualize how the points are distributed on a 2D plot. But we must bear in mind that Goiânia is located in a longitude between 49°10′ West and 49°20′ West, and in a latitude between 16°35′ South and 16°45′ South!

```python
plt.scatter(data["long"], data["lat"], alpha=.2)
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.show()
```

Whew. Apparently there are a lot of random points scattered far away from Goiânia. Let's take a more detailed look on these instances:

```python
data[~((data["lat"] <= -16.568957338589673)&\
    (data["lat"] >= -16.792978104243005)&\
    (data["long"] <= -49.15005619802892)&\
    (data["long"] >= -49.39739760249728))]
```

As far as I know, the addresses are reasonable, but for some reason the Bing Maps API wasn't able to identify the correct values for latitude and longitude. That's life. I will not change each value manually because it isn't scalable, so it wouldn't make sense in a context of full automated modeling. So I will impute `np.nan` for these cases.

```python
data.loc[~((data["lat"] <= -16.568957338589673)&\
    (data["lat"] >= -16.792978104243005)&\
    (data["long"] <= -49.15005619802892)&\
    (data["long"] >= -49.39739760249728)), ["lat", "long"]] = np.nan
```

Now let's try to see the scatter plot again, colored by the log(prices) of the apartments:

## Visualizing the variables together with the price of the apartments

Another useful way to verify if there are other absurd outliers is to plot the variables together with the log(price) of the apartments. We know beforehand that the price should increase if the variable increases. So if this "law" isn't followed, we may be suspicious. Let's plot all the possible combinations!

```python
import seaborn as sns

xs = ["condo_fee", "iptu", "numberOfRooms", "numberOfBathroomsTotal",
      "numberOfParkingSpaces", "floorSize", "floorLevel", "description_length",
      "total_cat"]

for x in xs:
    sns.regplot(x=data[x], y=np.log10(data["price"]), lowess=True, marker="x", line_kws=dict(color="r"))
    plt.xlabel(x)
    plt.ylabel("log10(apartment price)")
    plt.show()
```

From the plots above, I will impute np.nan on the following instances:

```python
data.loc[data["iptu"] >= 30000, "iptu"] = np.nan
data.loc[data["numberOfRooms"] >= 8, "numberOfRooms"] = np.nan
data.loc[(data["price"] <= 10**5.5)&(data["floorSize"] >= 400), "floorSize"] = np.nan
```

Let's visualize all the plots again:

```python
xs = ["condo_fee", "iptu", "numberOfRooms", "numberOfBathroomsTotal",
      "numberOfParkingSpaces", "floorSize", "floorLevel", "description_length",
      "total_cat"]

for x in xs:
    sns.regplot(x=data[x], y=np.log10(data["price"]), lowess=True, marker="x", line_kws=dict(color="r"))
    plt.xlabel(x)
    plt.ylabel("log10(apartment price)")
    plt.show()
```

It appears to be all OK now.

## Apartment main images

For the main images of the apartments, we should care if there is any "empty" image, as we automatically download them, and also for the dimensions of the images. Maybe there is some apartment without image too? Let's see:

```python
from PIL import Image

path = "C:/Users/cego/OneDrive/Área de Trabalho/Projetos Pessoais/Projeto, apartment prices v2/1_data_extraction/imgs/"

ids_found = []
x = []
y = []
c = []
ids_not_found = []
for ID in data["id"]:
    try:
        im = Image.open(path + ID + ".png")
        im = np.array(im)
        ids_found.append(ID)
        x.append(im.shape[0])
        y.append(im.shape[1])
        try:
            c.append(im.shape[2])
        except:
            c.append(1)
    except:
        ids_not_found.append(ID)
```

```python
len(ids_not_found)
```

There are 3 apartments without a main image downloaded. Let's drop these rows:

```python
data[data["id"].isin(ids_not_found)]
```

```python
data = data[~data["id"].isin(ids_not_found)].copy()
```

Now let's see if the dimensions of the images make sense:

```python
img_df = pd.DataFrame({"id": ids_found,
                      "x": x,
                      "y": y,
                      "c": c})

plt.scatter(img_df["x"], img_df["y"], c=img_df["c"], alpha=.2)
plt.colorbar()
plt.show()
```

Cerntaily a image with 0x0 dimensions doesn't make sense. Let's try to identify these apartments:

```python
img_df[img_df["x"] < 10]
```

There is only 1 apartment, so we can just drop the corresponding row:

```python
data = data[data["id"] != "288319718565533150857273231546188484051"]
```
